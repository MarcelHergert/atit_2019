```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

plt.set_cmap(None)
```

# CIFAR-10 Dataset

https://www.cs.toronto.edu/~kriz/cifar.html

```python
cifar = tf.keras.datasets.cifar10.load_data()[0]
image, labels = cifar[0], cifar[1]
```

```python
index = -1

_ = plt.imshow(image[index])
print(labels[index])
```

```python
index = 300

_ = plt.imshow(image[index])
print(labels[index])
```

```python
index = 1337

_ = plt.imshow(image[index])
print(labels[index])
```

# Exploration

```python
np.unique(labels, return_counts=True)
```

```python
heights, _ = np.histogram(labels)
plot_labels = np.arange(10) + 1
plt.bar(plot_labels, heights, tick_label=plot_labels)
```
